```python
import time
```

```python
import rdflib

start_time = time.time()
g = rdflib.Graph()
g.parse("data/amplus-stripped_without_edm_object-.nt")
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
units_query = """
SELECT DISTINCT ?dim_unit ?dim_value
WHERE {{
    {} <http://purl.org/collections/nl/am/dimension> ?dimension .
    ?dimension <http://purl.org/collections/nl/am/dimensionUnit> ?dim_unit .
    ?dimension <http://purl.org/collections/nl/am/dimensionValue> ?dim_value .
}}
"""

def create_dims_unit_dict(g, q, list_of_uris):
    entity_dims_dict = {}
    dim_units = []
    for uri in list_of_uris:
    #     print(uri, y)
        qres = g.query(q.format("<"+uri+">"))
        entity_dims_dict[uri] = {}
        for row in qres:
            try:
                entity_dims_dict[uri][row.dim_unit].append(row.dim_value)
            except:
                entity_dims_dict[uri][row.dim_unit] = [row.dim_value]
            dim_units.append(row.dim_unit)
    dim_units = list(set(dim_units))
    return entity_dims_dict, dim_units

start_time = time.time()
entity_dims_list_dict, dim_units = create_dims_unit_dict(g, units_query, list_of_uris)
print(len(dim_units))
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
def create_unit_count_dict(list_of_uris, unit_types, entity_dims_list_dict):
    unit_count_dict = {}
    for unit in unit_types:
        unit_count_dict[unit] = 0
    for entity in list_of_uris:
        for unit_type in unit_types:
            try:
#                 if len(entity_dims_dict[entity][unit_type])> 10:
#                     print(unit_type)
#                     print(entity_dims_dict[entity][unit_type])
                if len(entity_dims_list_dict[entity][unit_type]) > unit_count_dict[unit_type]:
                    unit_count_dict[unit_type] = len(entity_dims_list_dict[entity][unit_type])
            except:
                pass
    return unit_count_dict
```

```python
unit_count_dict = create_unit_count_dict(list_of_uris, dim_units, entity_dims_list_dict)
# dim_units_combined = dim_units
```

```python
entity_dims_list_dict_train, dim_units_train = create_dims_unit_dict(g, units_query, list_of_uris_train)
unit_count_dict_train = create_unit_count_dict(list_of_uris_train, dim_units_train, entity_dims_list_dict_train)
```

we remove units: 
- rdflib.term.Literal('x'), because there are only 2 occurances, and those are silly (XL, and 40½). 
- rdflib.term.Literal('-'), because there are only 2 occurances, and those are silly ('-'). 

All other units will be kept, and not collected, combined, because the are not (always) referering to the same dimensions of one objects (for example, "Cassette met 36 lepels" has (at least) two dimensions, of which I assume one is the weight of a single spoon, and the other is the weight of the combined object.

```python
def remove_unit(dim_units, unit):
    try:
        dim_units.remove(rdflib.term.Literal(unit))
        print("removed \'{}\'".format(unit))
    except:
        print("couldn't remove \'{}\'".format(unit))

remove_unit(dim_units, '-')
remove_unit(dim_units, 'x')
```

```python
def extend_dim_units(dim_units, unit_count_dict):
    new_dim_units = []
    for unit in dim_units:
        new_dim_units.append(unit)
        for i in range(1,unit_count_dict[unit]):
    #         print(unit,i+1)
            new_dim_units.append(unit+str(i+1))
    return new_dim_units
```

```python
def get_dim_by_unit(dims_dict, dim_unit, max_n_values, null_value=''):
    try:
        values = []
        for i in range(0, len(dims_dict[dim_unit])):
            values.append(dims_dict[dim_unit][i].replace(',','.')) #little bit of data cleaning
        values.sort(reverse=True)
        while len(values) < max_n_values: #fill with empty until as long as the longest with this unit
            values.append(null_value)
        return ','.join(values)
    except:
        return ','.join([null_value for i in range(0, max_n_values)])


def write_entities_for_units(list_of_uris, list_of_classes, dim_units, entity_dims_list_dict, unit_count_dict, new_fn):
    new_dim_units = extend_dim_units(dim_units, unit_count_dict)
    with open(new_fn, 'w') as f:
        f.write("entity,y,"+','.join(new_dim_units)+'\n')
        for entity, y in zip(list_of_uris, list_of_classes):
            list_of_values = []
            for dim_unit in dim_units:
                list_of_values.append(get_dim_by_unit(entity_dims_list_dict[entity], dim_unit, unit_count_dict[dim_unit]))
            f. write(entity+","+str(y)+','+','.join(list_of_values)+'\n')
```

```python
write_entities_for_units(list_of_uris, 
                         list_of_classes, 
                         dim_units, 
                         entity_dims_list_dict,
                         unit_count_dict,
                         "data/AM_entities_units_VALIDATION.csv")

write_entities_for_units(list_of_uris_train, 
                         list_of_classes_train, 
                         dim_units_train, 
                         entity_dims_list_dict_train,
                         unit_count_dict_train,
                         "data/AM_entities_units.csv")
```
